# Model Comparison

**This notebook records the overall model comparison for 3 sota models, and 5 datasets**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pyplot as plt
import numpy as np


datasets = ['Yelp2018', 'AmazonBook', 'Gowalla', 'MovieLens', 'GitStar']
models = ['LightGCN', 'UltraGCN', 'LayerGCN']

# Average Recall@20
recall_light = [0.063877, 0.041011, 0.182057, 0.257638, 0.044454]
recall_ultra = [0.067510, 0.067931, 0.186045, 0.277939, 0.053660]
recall_layer = [0.052600, 0.013900, 0.154100, 0.239300, 0.041600]

# Average NDCG@20
ndcg_light = [0.052329, 0.031566, 0.154196, 0.242705, 0.020112]
ndcg_ultra = [0.055522, 0.055337, 0.158005, 0.263404, 0.021617]
ndcg_layer = [0.041800, 0.010200, 0.123100, 0.228100, 0.016800]


x = np.arange(len(datasets))
width = 0.25

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), dpi=150)


def add_value_labels(ax, rects, fmt="{:.3f}", fontsize=9):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(
            fmt.format(height),
            xy=(rect.get_x() + rect.get_width() / 2, height),
            xytext=(0, 4),
            textcoords="offset points",
            ha='center',
            va='bottom',
            fontsize=fontsize
        )


rects1_1 = ax1.bar(x - width, recall_light, width,
                   label='LightGCN', color='#4C72B0', edgecolor='white')
rects1_2 = ax1.bar(x, recall_ultra, width,
                   label='UltraGCN', color='#55A868', edgecolor='white')
rects1_3 = ax1.bar(x + width, recall_layer, width,
                   label='LayerGCN', color='#C44E52', edgecolor='white')

ax1.set_ylabel('Recall@20 Score', fontsize=12, fontweight='bold')
ax1.set_title('Average Recall@20 Comparison Across Datasets',
              fontsize=14, fontweight='bold', pad=15)
ax1.set_xticks(x)
ax1.set_xticklabels(datasets, fontsize=11)
ax1.legend()
ax1.grid(axis='y', linestyle='--', alpha=0.7)

add_value_labels(ax1, rects1_1)
add_value_labels(ax1, rects1_2)
add_value_labels(ax1, rects1_3)


rects2_1 = ax2.bar(x - width, ndcg_light, width,
                   label='LightGCN', color='#4C72B0', edgecolor='white')
rects2_2 = ax2.bar(x, ndcg_ultra, width,
                   label='UltraGCN', color='#55A868', edgecolor='white')
rects2_3 = ax2.bar(x + width, ndcg_layer, width,
                   label='LayerGCN', color='#C44E52', edgecolor='white')

ax2.set_ylabel('NDCG@20 Score', fontsize=12, fontweight='bold')
ax2.set_title('Average NDCG@20 Comparison Across Datasets',
              fontsize=14, fontweight='bold', pad=15)
ax2.set_xticks(x)
ax2.set_xticklabels(datasets, fontsize=11)
ax2.legend()
ax2.grid(axis='y', linestyle='--', alpha=0.7)

add_value_labels(ax2, rects2_1)
add_value_labels(ax2, rects2_2)
add_value_labels(ax2, rects2_3)


plt.tight_layout(pad=3.0)

# Lưu ảnh nếu cần
# plt.savefig("metrics_comparison.png", bbox_inches="tight")

plt.show()
```
